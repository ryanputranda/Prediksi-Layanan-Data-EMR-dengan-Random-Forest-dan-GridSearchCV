```python
# prompt: hitung tiap data berdasarkan atribut cluster_encod untuk data layanan spesialis kompleks, layanan spesialis menengah dan layanan dasar ekonomis, buat plot barnya

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
#file_path = 'path/to/your/datasetclustering.xlsx'
#df = pd.read_excel(r'datasetclustering.xls')

# Count the number of data points for each cluster
cluster_counts = df['cluster'].value_counts()

# Create a bar plot
plt.bar(cluster_counts.index, cluster_counts.values)
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.title('Number of Data Points per Cluster')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding untuk kolom 'diagnosa'
le_diagnosa = LabelEncoder()
le_institusi = LabelEncoder()
le_area = LabelEncoder()
le_tipe = LabelEncoder()
le_biaya = LabelEncoder()
le_cluster = LabelEncoder()

df['diagnosa_encod'] = le_diagnosa.fit_transform(df['diagnosa'])
df['institusi_encod'] = le_institusi.fit_transform(df['institusi'])
df['area_encod'] = le_area.fit_transform(df['area'])
df['tipe_encod'] = le_tipe.fit_transform(df['tipe'])
df['biaya_berobat_encod'] = le_biaya.fit_transform(df['biaya_berobat'])
df['cluster_encod'] = le_cluster.fit_transform(df['cluster'])
```

```python
df
```

```python
# prompt: hapus kolom ke satu sampai ke enam

df = df.iloc[:, 6:]
df
```

```python
# Pisahkan fitur (X) dan target (y)
x = df[['diagnosa_encod', 'institusi_encod', 'area_encod', 'tipe_encod', 'biaya_berobat_encod']]
y = df[['cluster_encod']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
# prompt: analisa code saya, tambahkan hyperparameter tuning dengan gridsearch, sekalian atur param_gridnya dengan menganalisa datasetnya, model yang dipakai random forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define the random forest classifier
rfc = RandomForestClassifier()

# Define the hyperparameter grid
param_grid = {
  'n_estimators': [50, 100, 150],
  'max_depth': [5, 10, 15, 20, 25, 30],
  'min_samples_split': [2, 5, 10],
  'min_samples_leaf': [1, 2, 4]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(rfc, param_grid, cv=5)
grid_search.fit(x_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Train the model with the best hyperparameters
best_rfc = RandomForestClassifier(**grid_search.best_params_)
best_rfc.fit(x_train, y_train)

# Evaluate the model on the test set
y_pred = best_rfc.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestRegressor
#RF = RandomForestRegressor(n_estimators=50, max_depth=16, random_state=55, n_jobs=-1)
RF = RandomForestRegressor(max_depth= 15, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 100)
RF.fit(x_train, y_train)
```

```python
# Lakukan prediksi
y_pred = RF.predict(x_test)
```

```python
y_pred
```

```python
# prompt: round hasil y_pred
import numpy as np
y_pred = np.round(y_pred)
y_pred
```

```python
y_test
```

```python
# prompt: rubah tipe data y_pred dan y_test menjadi integer

y_pred = y_pred.astype('int')
y_test = y_test.astype('int')
```

```python
# Evaluasi model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# prompt: analisa semua code saya, tambahkan confusion matriks, buat visualisasinya

from sklearn.metrics import confusion_matrix
import seaborn as sns

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a heatmap visualization of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
# prompt: lakukan prediksi pada data testing 10 baris terakhir secara acak dan bandingkan antara atribut cluster aslinya dengan prediksi modelnya

# Select 10 random rows from the testing data
test_data = df.sample(n=10)

# Extract features and target labels
X_test = test_data[['diagnosa_encod', 'institusi_encod', 'area_encod', 'tipe_encod', 'biaya_berobat_encod']]
y_test = test_data['cluster_encod']

# Make predictions on the selected data
y_pred = RF.predict(X_test)

# Convert predicted clusters to original category names
y_pred_labels = le_cluster.inverse_transform(y_pred.astype(int))

# Extract original cluster labels for the selected data
y_test_labels = le_cluster.inverse_transform(y_test)

# Create a comparison table
comparison_df = pd.DataFrame({
    'Original Cluster': y_test_labels,
    'Predicted Cluster': y_pred_labels
})

# Display the comparison table
print(comparison_df.to_string())
```
